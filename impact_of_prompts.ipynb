```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
x1 = [68.9, 84.2, 95.0, 97.2, 94.0, 90, 85, 82]

# Market1501 plot
plt.plot(x, x1, marker='o', label='Market1501')

y = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y1 = [40.3, 80.5, 94.5, 96.0, 95.1, 88, 84, 83]

# DukeMTMC plot
plt.plot(y, y1, marker='s', linestyle='-.', label='DukeMTMC')

plt.xlabel("Number of prompts")
plt.ylabel("Rank-1 accuracy")
plt.title("Impact of number of prompts pooled")
plt.legend()  # Add this line for the legend
#plt.grid(True)  # Optional: makes it cleaner
plt.tight_layout()
plt.show()
```
